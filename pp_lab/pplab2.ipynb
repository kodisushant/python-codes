```python
#to display the diagonal elements
r=int(input("Enter the no: of rows:"))
c=int(input("Enter the no: of cols:"))
m=[]
if r==c:
    for i in range(r):
        row=[]
        for j in range(c):
            row.append(int(input()))
        m.append(row)
    print("Matrix:\n[")
    for i in range(r):
        for j in range(c):
            print(m[i][j],end=" ")
        print()
    print("]")
    print("Diagonal elements:",end = " ")
    for i in range(r):
        for j in range(c):
            if i==j:
                print(m[i][j],end=" ")
else:
    print("Its is not a Square Matrix")
```

```python
#to display the transpose of a matrix
r=int(input("Enter the no: of rows:"))
c=int(input("Enter the no: of cols:"))
m=[]
if r==c:
    for i in range(r):
        row=[]
        for j in range(c):
            row.append(int(input()))
        m.append(row)
    print("Matrix:\n[")
    for i in range(r):
        for j in range(c):
            print(m[i][j],end=" ")
        print()
    print("]")
    print("Transpose of the matrix:\n[")
    for i in range(r):
        for j in range(c):
            print(m[j][i],end=" ")
        print()
    print("]")
else:
    print("Its is not a Square Matrix")
```

```python
# sum of matrix
r1=int(input("Enter the no: of rows:"))
c1=int(input("Enter the no: of cols:"))
m1=[]
r2=int(input("Enter the no: of rows:"))
c2=int(input("Enter the no: of cols:"))
m2=[]
if r1==c1==r2==c2:
    print("enter matrix 1:")
    for i in range(r1):
        row=[]
        for j in range(c1):
            row.append(int(input()))
        m1.append(row)
    print("Enter matrix 2:")
    for i in range(r2):
        row=[]
        for j in range(c2):
            row.append(int(input()))
        m2.append(row)
    print("Matrix 1 :\n[")
    for i in range(r1):
        for j in range(c1):
            print(m1[i][j],end=" ")
        print()
    print("]")
    print("Matrix 2 :\n[")
    for i in range(r2):
        for j in range(c2):
            print(m2[i][j],end=" ")
        print()
    print("]")

    m3=[]
    for i in range(r1):
        row=[]
        for j in range(c1):
            row.append(m1[i][j]+m2[i][j])
        m3.append(row)
    print("Matrix 3 :\n[")
    for i in range(r2):
        for j in range(c2):
            print(m3[i][j],end=" ")
        print()
    print("]")
else:
    print("It is not possible")
```
