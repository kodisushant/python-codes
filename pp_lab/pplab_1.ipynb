```python
#to display a square matrix
r = int(input("Enter no: of rows:"))
c = int(input("Enter no: of cols:"))
mat = []
if r==c:
    for i in range(r):
        row = []
        for j in range(c):
            row.append(int(input()))
        mat.append(row)
    print("Square Matrix:")
    print("[",end="")
    for i in range(r):
        for j in range(c):
            print(mat[i][j],end=" ")
        print()
    print("]",end=" ")
else:
    print("Not a Square matrix")
```
