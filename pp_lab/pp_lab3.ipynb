```python
import numpy as np
from numpy.linalg import det 
mat1 = np.array([[1,2,3],[4,5,6],[7,8,1]])
print("Matrix:\n",mat1)
print("Determinant of mat1:",end=" ")
det1 = det(mat1)
print('%d'%det1)
```

```python
import numpy as np
from numpy.linalg import inv 
mat1 = np.array([[1,2,3],[4,5,6],[7,8,1]])
print("Matrix:\n")
print("Inverse of mat1:")
inv1 = inv(mat1)
print(inv1)
```
