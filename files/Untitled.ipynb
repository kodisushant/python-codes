```python
with open("marks.txt","r+") as n:
    g = n.readlines() 
    j=1
    for i in n.readlines(): 
        if i.split()[-1] == 'f' and int(i.split()[-2]) >60:
            print(f"Lines {j:<2}",i," gold",end="")
        elif i.split()[-1] == 'f' and int(i.split()[-2]) <60:
            print(f"Lines {j:<2}",i," bronze",end="")
        elif i.split()[-1] == 'm' and int(i.split()[-2]) >60:
            print(f"Lines {j:<2}",i," silver",end="")
        elif i.split()[-1] == 'm' and int(i.split()[-2]) <60:
            print(f"Lines {j:<2}",i," plastic",end="")
```

```python
with open("marks.txt","r") as n:
    edited = []
    i = n.readline()
    while i != "": 
        if i.split()[-1] == 'f' and int(i.split()[-2]) >60:
            i=i[:-1]+" GOLD"
        elif i.split()[-1] == 'f' and int(i.split()[-2]) <60:
            i=i[:-1]+" BRONZE"
        if i.split()[-1] == 'm' and int(i.split()[-2]) >60:
            i=i[:-1]+" SILVER"
        elif i.split()[-1] == 'm' and int(i.split()[-2]) <60:
            i=i[:-1]+" STEEL"
        edited.append(i)
        i=n.readline()
print(edited)
# edited.join("\\n")
with open("first.txt","w") as n:
    n.write("\n".join(edited))
```
