```python
x = int(input("Enter 1st angle:"))
y = int(input("Enter 2nd angle:"))
z = int(input("Enter 3rd angle:"))
if(x+y+z==180):
    if(x==y==z):
        print("Equilateral triangle")
    elif(x==90 or y==90 or z==90):
        print("Right angle triangle")
    else:
        print("Normal triangle")
else:
    print("Triangle cannot be formed")
```

```python
key=1
while key:
    num1 = float(input("Enter 1st number:"))
    opr = input("Enter any operator(+,-,*,/,**(exponent),1Q1(to exit)")
    num2 = float(input("Enter 2nd number:"))
    res=0
    match opr:
        case "+":
            res=num1+num2
            print(res)
        case "-":
            res=num1-num2
            print(res)
        case "*":
            res=num1*num2
            print(res)
        case "/":
            if num2==0:
                print("Error")
            else:
                res=num1/num2
                print(res)
        case "**":
            res=num1**num2
            print(res)
        case "Q":
            print("Out of calculator")
            key=0
        case _:
            print("Error")
```

```python
account_type = input("Enter account type (saving(S)/checkbook(C)): ")

if account_type == "C":
    min_balance = 1000
elif account_type == "S":
    min_balance = 0

balance = float(input("Enter initial balance: "))

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Show Balance")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    match choice:
        case 1:
            amount = float(input("Enter deposit amount: "))
            if amount > 0:
                balance += amount
                print("Amount deposited.")
            else:
                print("Invalid amount.")

        case 2:
            amount = float(input("Enter withdrawal amount: "))
            if amount > 0 and balance - amount >= min_balance:
                balance -= amount
                print("Amount withdrawn.")
            else:
                print("Minimum balance required")

        case 3:
            print("Current Balance:", balance)

        case 4:
            print("Exiting program.")
            break

        case _:
            print("Invalid choice.")
```

```python
n=int(input("Enter a number:"))
i=1
sum=0
while i<=n:
    sum+=i
    i+=1
print("Sum =",sum)
sum1=0
for j in range(1,n+1):
    sum1+=j
    j+=1
print("Sum =",sum1)
```
