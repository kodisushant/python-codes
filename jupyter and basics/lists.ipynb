```python
print("Quadratic eq: ax^2 + bx +c = 0")
a=float(input("Enter value of a:"))
b=float(input("Enter value of b:"))
c=float(input("Enter value of c:"))
d= b**2 - 4*a*c
x1=0
x2=0
if(d>=0):
    print("Real roots")
    x1=(-b+d**(1/2))/(2*a)
    x2=(-b-d**(1/2))/(2*a)
    print(x1,x2)
else:
    print("Imaginary roots")
    x1=(-b+d**(1/2))/(2*a)
    x2=(-b-d**(1/2))/(2*a)
    print(x1,x2)
```
