```python
ops = {
    '+': lambda a, b: a+b,
    '-': lambda a, b: a-b,
    '*': lambda a, b: a*b,
    '/': lambda a, b: a/b,
}

def calculate(a, op, b):
    return ops[op](a, b)

print(calculate(10, '+', 5))   # 15
print(calculate(10, '-', 3))   # 7
print(calculate(4,  '*', 6))   # 24
print(calculate(15, '/', 4))   # 3.75
```

```python
'''
Task  
•  Start with a list of product prices.
•  Apply 10% discount using map().
•  Then add 18% GST on the discounted price using another map().
•  Finally round to 2 decimal places.
•  Print original, discounted, and final prices side by side.
'''
prices = [1000, 250, 4500, 799, 120]
dis = list(map(lambda x : x-x*0.1,prices))
tax = list(map(lambda y : y+y*0.18,dis))
f = list(map(lambda z: round(z,2),tax))
print(f'{'Original':<10} {'Discounted':<12} {'Final':<10}')
for p, d, f in zip(prices, dis, f):
    print(f'{p:<10} {d:<12.2f} {f:<10.2f}')
```

```python
def is_Prime(n): 
    v=0 
    if n<2: 
        return False
    else:
        for i in range(2,int(n**0.5)+1): 
            if n%i==0:
                v+=1
            else: 
                v+=0
    if v>0:
        return False
    else: 
        return True
nums = list(range(2,51))
primes = list(filter(is_Prime,nums)) 
print("Primes:",primes)

twin_primes = [(p,p+2) for p in primes if is_Prime(p) and is_Prime(p+2)]
print("Twin Primes:",twin_primes)
```

```python
''' 
•  Sort the product list by price (ascending).
•  Sort by category alphabetically, then by price descending within each category.

'''
products = [
    {'name': 'Laptop',   'category': 'Electronics', 'price': 55000},
    {'name': 'T-Shirt',  'category': 'Clothing',    'price': 799},
    {'name': 'Phone',    'category': 'Electronics', 'price': 25000},
    {'name': 'Jeans',    'category': 'Clothing',    'price': 1999},
    {'name': 'Tablet',   'category': 'Electronics', 'price': 18000},
    {'name': 'Jacket',   'category': 'Clothing',    'price': 3499},
]
#by price 
by_price = sorted(products, key=lambda x:x['price'])
for p in by_price:
    print(f"{p['name']:10} {p['price']}")

#by alphabetically of category and then descending 
by_cat_price = sorted(products, key = lambda x:(x["category"],-x["price"]))
for p in by_cat_price:
    print(f"{p['category']:12} {p['name']:10} {p['price']}")
```

```python
'''  Task  
•  Given a sentence, count the frequency of each unique word (case-insensitive).
•  Use list comprehension to build the frequency dictionary.
•  Then use sorted() + list comprehension to display top 5 most frequent words.
'''
text = '''python is great python is easy
python helps you learn great things easily
you can learn python easily and quickly'''
words = text.lower().split() 
freq = {word: words.count(word) for word in set(words)}
print(freq)
top_5 = sorted(freq.items(),key= lambda x:(-x[1],x[0],-len(x[0])))[:5]
print()
print("Top 5:")
for w,c in top_5:
    print(f"{w:<12} {c}")
```

```python
'''
Exercise 17: Random Password Generator   ★★
  Task  
•  Generate a random password of given length.
•  Must include at least 1 uppercase, 1 lowercase, 1 digit, 1 special character.
•  Use random.shuffle() to randomize the character order.
  Starter Code  
'''
import random , string

def generate_password(length = 12):
    upper = random.choices(string.ascii_uppercase,k=2)
    lower = random.choices(string.ascii_lowercase,k=2)
    digits = random.choices(string.digits,k=2)
    special = random.choices('!@#$%^&*',             k=2)

    all_chars = string.ascii_letters + string.digits + '!@#$%^&*'
    extra = random.choices(all_chars, k=length - 8)

    password = upper + lower + digits + special + extra
    random.shuffle(password)
    return " ".join(password)

random.seed(42)
for i in range(3):
    print(f'Password {i+1}: {generate_password(12)}')
```

```python
'''
Exercise 19: Student Report Card Generator   ★★
  Task  
•  Use filter() to get students who passed all subjects (min score >= 40).
•  Use map() to compute each student's average.
•  Use sorted() to rank by average descending.
•  Use list comprehension to format the final report.

'''
students = [
    {'name': 'Ananya', 'scores': [78, 82, 91, 67]},
    {'name': 'Balram', 'scores': [55, 38, 72, 60]},  # fails (38 < 40)
    {'name': 'Chitra', 'scores': [88, 95, 90, 92]},
    {'name': 'Dinesh', 'scores': [40, 45, 52, 48]},
    {'name': 'Ekalavya','scores':[72, 68, 75, 80]},
]
passed = list(filter(lambda x:min(x["scores"])>=40,students))

avg = list(map(lambda x:{**x, "average": sum(x["scores"])/4},passed))

rank_wise = sorted(avg,key=lambda x:-x["average"])

print(f"{'Rank':<6}{'Name':<12}{'Avg':>6}")
print('-' * 26)
for i, s in enumerate(rank_wise, 1):
    print(f"{i:<6}{s['name']:<12}{s['average']:>6.1f}")
```

```python
'''
Exercise 20: Text Analyser (Mini NLP)   ★★★
  Task  
•  Write a function analyse(text) that uses map, filter, sorted & list comprehension.
•  Remove punctuation, convert to lowercase, split into words.
•  Filter stop words (the, is, a, an, and, in, of, to, for, it).
•  Use map() to stem words (remove trailing 's' or 'ing').
•  Build sorted word frequency dictionary and return top N words.
'''
import re,string

STOP_WORDS = {'the','is','a','an','and','in','of','to','for','it','this'}

def stem(word):
    if word.endswith('ing') and len(word) > 5:
        return word[:-3]
    if word.endswith('s') and len(word) > 3:
        return word[:-1]
    return word

def text_analyser(text,top_n=5):
    # Step 1: Clean — remove punctuation, lowercase
    clean = re.sub(r'[^\w\s]', '', text.lower())
    words = clean.split()

    # Step 2: Filter stop words
    filtered = list(filter(lambda x:x not in STOP_WORDS, words))

    # Step 3: Stem with map
    stemmed = list(map(lambda x:stem(x), filtered))

    # Step 4: Frequency dict
    freq = {w: stemmed.count(w) for w in set(stemmed)}

    # Step 5: Top N
    top = sorted(freq.items(), key=lambda x:-x[1])[:top_n]
    return top

text = '''Python is changing the world of programming.
          Programming with Python teaches you creative thinking.
          Creative coders use Python for building great systems.'''

results = text_analyser(text)
print('Top words:')
for word, count in results:
    print(f'  {word:<15} {count}')
```

```python
import random

def estimate_pi(n_points):
    random.seed(0)
    inside = 0
    for _ in range(n_points):
        x = random.random()        
        y = random.random()
        if x**2 + y**2 <= 1:       
            inside += 1
    return 4 * inside / n_points

for n in [100, 1000, 10000, 100000]:
    pi_est = estimate_pi(n)
    print(f'N={n:>8}: π ≈ {pi_est:.5f}  (error: {abs(pi_est - 3.14159):.5f})')
```
