```python
###Update marks of a student if  student exists.###
stu = {"ram":97,"sita":99,"laxman":90}
s = input("Enter student name:")
if s in stu:
    stu[s]=int(input("update the marks:"))
    print("Updated marks:",stu)
else:
    print("does not exist")
# for k in stu:
#     if s==k:
#         stu[k]=int(input("update the marks:"))
#     elif:
#         print("name does not exist")
```

```python
###Delete a contact from STUDENTdictionary.###
stu = {123:"shnt",124:"abhi",125:"bha",126:"susl"}
x = int(input("Enter a number to delete:"))
if x in stu:
    stu.pop(x)
    print(stu)
else:
    print("number does not exist")
```

```python
###Count number of items in a dictionary.###
stu = {123:"shnt",124:"abhi",125:"bha",126:"susl"}
print(len(stu))
```

```python
###update Student dictionary by adding marks  and find sum of all values.###
stu = {"ram":97,"sita":99,"laxman":90}
for k in stu.keys():
    stu[k]+=10
stu
sum("the sum is",stu.values())
```

```python
###Check whether a key exists in dictionary.###
stu = {"ram":97,"sita":99,"laxman":90}
x = input("enter a key to check:")
if x in stu:
    print(x,"exists")
else:
    print(x,"does not exists")
```

```python
###Print all keys and all values separately.###
stu = {"ram":97,"sita":99,"laxman":90}
for k,v in stu.items():
    print(k,v)
```

```python
###Add a new item to dictionary.###
stu = {"ram":97,"sita":99,"laxman":90}
n = int(input("how many elements to add in the dict:"))
for i in range(n):
    x=input("enter the key:")
    stu[x]=int(input("enter value:"))
stu
```

```python
###Remove all items from dictionary.###
stu = {"ram":97,"sita":99,"laxman":90}
stu.clear()
stu
```

```python
###Merge two dictionaries using the | merge operator.###
stu1 = {"ram":97,"sita":99,"laxman":90}
stu = {123:"shnt",124:"abhi",125:"bha",126:"susl"}
stu2 = stu1 | stu
stu2
```

```python
###sort on the item with highest price###
bill = {"maggi":50,"biryani":200,"water":20,"machuria":100,"onions and soamp":0}
sort_bill = dict()
# for val in sorted(bill.values()): 
#     for k in bill:
#         if bill[k]==val:
#             sort_bill[k]=val
# print(sort_bill)
sort_bill = dict(sorted(bill.items() , key = lambda x:x[1],reverse=True))
print(sort_bill)
for k,v in sort_bill.items():
    print(f"{k:20}: ${v}")
```

```python
###Merge two product dictionaries into a new dictionary using ** unpacking.###
stu1 = {"ram":97,"sita":99,"laxman":90}
stu = {123:"shnt",124:"abhi",125:"bha",126:"susl"}
stu2 = {**stu1,**stu}
stu2
```
