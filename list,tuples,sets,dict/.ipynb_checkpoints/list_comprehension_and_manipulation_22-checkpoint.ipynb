```python
'''📘 Concept:  Lists are mutable sequences. You build them using append() to add to the end, and insert(i, val) to add at a specific position.
You remove with remove(val) (first match) or pop(i) (by index).
Swapping two elements in-place uses a temporary variable — the classic pattern for in-place reversal and rearrangement.
Q1.	Start with an empty list. Using a for loop, add numbers from 1 to 10 using append(). Then use a while loop with remove() to delete all even numbers.
Expected: [1, 3, 5, 7, 9]
Hint: Build a separate list of evens first, then loop and remove each. Avoid modifying a list while iterating over it.
'''
#li = [i for i in range(1,11)]
li = []
for i in range(1,11):
    li.append(i)
j=1
while j<=10:
    if j%2==0:
        li.remove(j)
    j+=1
print(li)
```

```python
'''
Q2.	Given lst = [10, 20, 30, 40, 50], insert the value 25 at the correct position so the list stays sorted. Use insert() only — no sorting allowed.
Then insert 5 and 55 and verify order is maintained.
Expected: [5, 10, 20, 25, 30, 40, 50, 55]
Hint: Loop through to find the first index where lst[i] > value, then insert(i, value).
'''
lst = [10, 20, 30, 40, 50]
num = int(input())
for i in range(len(lst)):
    if num < lst[i]:
        lst.insert(i,num)
        break
    elif num > lst[len(lst)-1]:
        lst.insert(len(lst),num) 
print(lst)
```

```python
'''
Q3.	Given lst = ['a', 'b', 'c', 'd', 'e'], reverse it in-place using a loop and index swapping. Do NOT use reverse(), reversed(), or slicing.
Expected: ['e', 'd', 'c', 'b', 'a']
Hint: Use two pointers: left=0, right=len-1. Swap using a temp variable, then move both inward.
'''
lst = ['a', 'b', 'c', 'd', 'e'] 
for i in range(len(lst)//2):
    left = i 
    right = len(lst) -i- 1
    temp = lst[left]
    lst[left] = lst[right]
    lst[right] = temp 
    
print(lst)
```

```python
''' 
Q4.CHALLENGE: Given lst = [3, 1, 4, 1, 5, 9, 2, 6, 5], remove all duplicate values keeping only the first occurrence of each.
Do it using a loop, a seen list, and append() — no sets allowed.
Expected: [3, 1, 4, 5, 9, 2, 6]
Hint: Maintain a seen list. For each element, check if it is already in seen using 'in'. If not, append to both seen 
'''
lst = [3, 1, 4, 1, 5, 9, 2, 6, 5]
# for i in lst:
#     if lst.count(i)>1: 
#         lst.remove(i)
# print(lst)
seen = []
for i in lst:
    if i not in seen:
        seen.append(i)
print(seen)
```

```python
'''
Q5.	Given nums = [4, 7, 2, 9, 1, 5, 8, 3, 6], find the second largest and second smallest values using a single loop. No sorting allowed.
Expected: second largest = 8,  second smallest = 2
Hint: Track first and second largest separately. Update second when a new value is between the two extremes
'''
# nums = [9, 8, 7]
nums = [4, 7, 2, 9, 1, 5, 8, 3, 6] 
maxx = max2 = float('-inf')
minn = min2 = float('inf')
for i in nums:
    if i > maxx:
        max2 = maxx 
        maxx = i 
    elif i > max2:
        max2 = i
    if i < minn:
        min2 = minn 
        minn = i
    elif i < min2:
        min2 = i  
print(max2 , min2)
```

```python
'''
Q6.	Given lst = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5], count the frequency of each element using a loop and a plain dictionary (no Counter).
Then find the most frequent element.
Expected: {3:2, 1:2, 4:1, 5:3, 9:1, 2:1, 6:1}  →  most frequent = 5
Hint: freq[x] = freq.get(x, 0) + 1. Then loop over freq.items() to find the key with the highest value.
'''
lst = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
# lst2 = list(set(lst))
# freq = dict()
# for i in lst2:
#     freq[i]=lst.count(i)
# print(freq)
freq = {}
for x in lst:
    freq[x] = freq.get(x, 0) + 1       # get current count (default 0), add 1

# maxx = max(freq.items(), key = lambda x:x[1]) 
# print(maxx)
maxx = max(freq.values())
for k , v in freq.items(): 
    if v == maxx:
        print("the most repeated no: is",k,"whose frequency is",v)
```

```python
''' 
Q7.Given lst = [10, 20, 30, 40, 50, 60], find all indices where the value is greater than 25. Use a loop — no index() or list comprehension.
Expected: [2, 3, 4, 5]
Hint: Use enumerate(lst) to get both index and value. Append the index to a result list when the condition is met.
'''
lst = [10, 20, 30, 40, 50, 60]
res = []
for ind , val in enumerate(lst): 
    if val > 25: 
        res.append(ind)
print(res)
''' 
lst = [10, 20, 30, 40, 50]

for index, value in enumerate(lst):
    print(index, value)

# Output:
# 0  10
# 1  20
# 2  30
# 3  40
# 4  50
'''
```

```python
''' 
Q8.CHALLENGE: Given lst = [5, 3, 8, 1, 9, 2, 7, 4, 6], find the longest increasing run (consecutive elements where each is larger than the previous). 
Return the start index, end index, and the sublist.
Expected: start=4, end=6, sublist=[9] — wait, longest is [1,9]? Check all runs carefully.
Hint: Track current run length and best run. When lst[i] <= lst[i-1], the run resets. Store the best start/end indices.
'''
lst = [5, 3, 8, 1, 9, 2, 7, 4, 6]

# Track current run
curr_start = 0
curr_len   = 1

# Track best run seen so far
best_start = 0
best_len   = 1

for i in range(1, len(lst)):
    if lst[i] > lst[i-1]:           # run continues
        curr_len += 1
    else:                           # run breaks
        curr_start = i              # reset start to current index
        curr_len   = 1              # reset length

    if curr_len > best_len:         # is this run the best so far?
        best_len   = curr_len
        best_start = curr_start

best_end = best_start + best_len - 1
sublist  = lst[best_start : best_end + 1]

print("start   =", best_start)   # 1
print("end     =", best_end)     # 2
print("sublist =", sublist)      # [3, 8]
```

```python
''' 
Q9.	Given nums = [1, 2, 3, 4,5],build a new list where each element is the running total (cumulative sum) up to that position using a loop and append().
Expected: [1, 3, 6, 10, 15]
Hint: Keep a running variable total = 0. Each iteration: total += nums[i]; result.append(total).
'''
nums = [1, 2, 3, 4,5] 
sum_list = []
total = 0
for i in nums:
    total+=i 
    sum_list.append(total)
print(sum_list)
```

```python
''' 
Q10.Given lst = [5, 3, 8, 1, 9, 2, 7], compute the average using a loop (no sum()). Then build two new lists in a second loop: above_avg and below_avg.
Expected: avg=5.0   above=[5, 8, 9, 7]  below=[3, 1, 2]  (equal values can go in either)
Hint: First loop: accumulate total and count. Second loop: compare each value to avg and append to the correct list.
'''
lst = [5, 3, 8, 1, 9, 2, 7] 
total = 0
for i in lst:
    total+=i 
avg = total/len(lst)
abv = []
blw = []
for i in lst:
    if i >= avg:
       abv.append(i)
    else:
        blw.append(i)
print("above average =",abv) 
print("below average =",blw)
```

```python
''' 
Q11.Given prices = [100, 250, 80, 430, 150], apply a 10% discount to prices above 200 and a 5% discount to the rest. 
Build a new discounted list using a loop.
Expected: [95.0, 225.0, 76.0, 387.0, 142.5]
Hint: new_price = price * 0.90 if price > 200 else price * 0.95. Append each new_price to result.
'''
prices = [100, 250, 80, 430, 150]
disc = []
for i in prices: 
    new_price = i * 0.90 if i > 200 else i * 0.95
    disc.append(new_price)
print(disc)
```

```python
''' 
CHALLENGE: Given nums = [2, 5, 1, 8, 3, 9, 4, 7, 6], build a new list where each element is replaced by the maximum value seen so far (running maximum).
Then build a second list of the running minimum.
Expected: running_max = [2,5,5,8,8,9,9,9,9]   running_min = [2,2,1,1,1,1,1,1,1]
Hint: running_max = nums[0]. For each element, running_max = max(running_max, nums[i]) — but implement max() with an if statement, no built-ins.
'''
nums = [2, 5, 1, 8, 3, 9, 4, 7, 6] 
run_max = []
run_min = []
maxx = nums[0]
minn = nums[0]
for i in nums: 
    if i > maxx:
        maxx = i
    if i < minn: 
        minn = i 
    run_max.append(maxx)
    run_min.append(minn)
print(run_max)
print(run_min)
```

```python
''' 
Q13.Given scores = [85, 92, 78, 95, 88, 72, 91], compute mean, median, and mode using only loops and a dictionary. 
Do NOT use the statistics module or sum().
Formulas: mean = total/count   median = sorted_lst[n//2]   mode = key with max freq
Expected: mean=85.86  median=88  mode=no repeat so pick highest freq
Hint: Mean: loop to accumulate total. Median: sort (allowed here), pick middle index. Mode: build freq dict, find key with max value.
'''
scores = [85, 92, 78, 95, 88, 72, 91] 
total =0
for i in scores:
    total+=i 
mean = total/len(scores)
scores.sort()
median = scores[len(scores)//2]
print(f"Mean {mean:.2f}")
print("Median:",median)
```
