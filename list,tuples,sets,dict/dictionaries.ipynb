```python
score = {"Arjun":45,"Priya":56,"Ravi":23}
for name in score:
    print(name)
print()
for scores in score.values():
    print(scores)
print()
for details in score.items():
    print(details)
print()
topper = max(score,key=score.get)#checks max by comparing value
print(topper)
print(max(score))#check the alphabetical order of keys
print()

sortt_dict = dict(sorted(score.items()))#sorts based on alphabetical order of keys
print(sortt_dict)
print()

sortt_vales = {}
for value in sorted(score.values()):#here sorted(score.values()) creates a sep list and stores only values in ascending order(see last 2nd line)
    for key in score:
        if score[key]==value:
            sortt_vales[key]=value
print(sortt_vales)
print(sorted(score.values()))
print()
```
